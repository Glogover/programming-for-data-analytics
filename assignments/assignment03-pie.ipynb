```python
import pandas as pd # Importing the pandas library for data manipulation and analysis
```

```python
import matplotlib.pyplot as plt # Importing matplotlib for data visualization
```

```python
url = "https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download" # URL of the CSV file
```

```python
df = pd.read_csv(url) # Reading the CSV file from the URL into a Pandas DataFrame
#df.head() # Displaying the first few rows of the DataFrame
```

```python
df['domain'] = df['Email'].str.split('@').str[1] # Extracting the domain from the email addresses
#df.head(3) # Displaying the first three rows to verify the new 'domain' column
```

```python
counts = df['domain'].value_counts() # Counting the occurrences of each domain
```

```python
#counts # Displaying the counts of each domain
```

```python
plt.figure(figsize=(8, 8)) # Setting the figure size for the pie chart
plt.pie( # Creating the pie chart
    counts, # Data for the pie chart
    labels=counts.index, # Labels for each slice
    autopct='%1.1f%%', # Displaying the percentage on each slice
    startangle=90, # Starting angle of the pie chart      
    counterclock=False,  # Drawing the pie chart in a clockwise direction
    shadow=True,    # Adding a shadow effect to the pie chart    
    wedgeprops={'edgecolor': 'white', 'linewidth': 1},  # Styling the wedges with white edges
    colors=plt.cm.Set3.colors # Using a colormap for slice colors
)

plt.title('Pie chart of email domains', fontsize=16, fontweight='bold', pad=20) # Adding a title to the pie chart
plt.tight_layout() # Adjusting the layout to fit everything nicely
plt.show() # Displaying the pie chart
```
